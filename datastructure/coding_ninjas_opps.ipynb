```python
class ExpenseTracker:
    
    def __init__(self,category,balance,budget):

    #public attribute
        self.category = category

    # private attribute
        self.__balance = balance
        self.budget = budget
        
    def get_private_attributes(self):
        print(self.__balance)
        


  

    def __get_balance(self):
        print(self.__balance)   
    #     print(self._ExpenseTracker__get_balance())   # this will call the function continuouly i.e. recursion and the depth will be exceeded
    @staticmethod
    def capital(string):
        return string.capitalize()

    @staticmethod
    def floats(number):
        return float(number)

    @classmethod
    def objectcreator(cls,diary_entry:str):
        category,balance,budget = diary_entry.split(" ")

        return ExpenseTracker(cls.capital(category),   
                   cls.floats(balance),
                   cls.floats(budget))
            
```

```python
# -----------------PARENT CLASS---------------
class vehicle:
  def __init__(self,make, model, fuel):
    self.make = make
    #private attributes
    self.__model = model
    self.__fuel = fuel

  def __private_method(self):
    print("The is in private method in the parent class")
    print(self.make)

  def show_privatemethods(self):
    print("hie i am a method in parent class")
    print(vehicle.__model,vehicle.__fuel)


# -----------------child class--------------------

class car(vehicle):
  def __init__(self, make ,model,fuel,ac,sunroof):
    # parent attributes
    vehicle.make = make
    vehicle.__model = model
    vehicle.__fuel = fuel
    # child attributes
    self.ac = ac
    self.sunroof = sunroof

  def get_parent_attribute(self):
    print(vehicle.make,vehicle.__model,vehicle.__fuel,sep=" ")

  def get_privatemethod_ofparent(self):
    self._vehicle__private_method()
    print("i am private of private")
```

# operators overloading

```python
class vehicle:
    def __init__(self,make,model,fare):
        self.make = make
        self.model = model
        self.fare = fare
    def __str__(self):
        return f"i am a car of model {self.make} of the year {self.model} and the fare is {self.fare} rupees/min."
    
    def __add__(self,other):  # self is the 1st object and 'other' is the second object
        return self.fare + other.fare
    
    def __sub__(self,other):
        return self.fare - other.fare
    
    def __mul__(self,other):
        return self.fare * other.fare
    
    def __truediv__(self,other):
        return self.fare / other.fare
    
    def __and__(self,other):
        return self.fare & other.fare
```

# abstract class


Abstract classes is basically a interface, which states which methods should be bydefault there in the child classis
eg agar parent mai get_path function hai tho child mai bhi hona hi chahiye nahitho child class nahi banega                                                                                                                                                     
Agar parent mai __init__ hai tho child mai __init__ nahi dena padega

```python
from abc import ABC , abstractmethod

class TextReaderAbstract(ABC):
    
    def __init__(self,path,filename):
        self.path = path
        self.filename = filename
    
    @abstractmethod
    def get_path(self):
        pass
    
    @abstractmethod
    def get_filename(self):
        pass
    
class TextReader(TextReaderAbstract):
    
    def get_path(self):
        return self.path
    def get_filename(self):
        return self.filename
    
    
ob = TextReader("desktop/ml","datastructure")
```
